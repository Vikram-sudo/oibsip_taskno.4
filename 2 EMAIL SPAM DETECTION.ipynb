# EMAIL SPAM DETECTION WITH MACHINE LEARNING

```python
df.drop(columns=['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df=df.rename(columns={'v1':'Target' , 'v2':'Message'})
df.head()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(keep='first',inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.size
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
df['Target'] = encoder.fit_transform(df['Target'])
print(df['Target'])
```

```python
df.head() #Labeled Spamed as 1 and ham as 0
```

```python
#Visualisation of data
# Count the occurrences of each encoded value
value_counts = df['Target'].value_counts()

# Create the pie chart
plt.pie(value_counts, labels=['ham', 'spam'], autopct="%0.2f%%")

# Display the chart
plt.show()
```

```python
x=df['Message']
y=df['Target']
```

```python
print(x)
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train, y_test= train_test_split(x,y, test_size=0.2, random_state=3)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer()
# CountVectorizer can transform raw text data into numerical feature vectors 
```

```python
x_train_cv=cv.fit_transform(x_train)
x_test_cv = cv.transform(x_test)
```

```python
print(x_train_cv)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(x_train_cv,y_train)
prediction_train=lr.predict(x_train_cv)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_train,prediction_train)*100)
```

```python
prediction_test=lr.predict(x_test_cv)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,prediction_test)*100)
```

# Support Vector Machine

```python
from sklearn.svm import SVC
svm=SVC()
```

```python
svm.fit(x_train_cv,y_train)
prediction_train2=svm.predict(x_train_cv)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_train,prediction_train2)*100)
```

```python
prediction_test2=svm.predict(x_test_cv)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,prediction_test2)*100)
```
